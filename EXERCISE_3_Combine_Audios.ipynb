```python
import glob
import sys
import os
from pydub import AudioSegment
```

```python
includeDir = "/content/drive/MyDrive/VoiceClone/Experiment_Audio/SciAudio/"
ST_0 = AudioSegment.from_mp3(includeDir + "scientist-0.mp3")
ST_1 = AudioSegment.from_mp3(includeDir + "scientist-1.mp3")
ST_2 = AudioSegment.from_mp3(includeDir + "scientist-2.mp3")
ST_3 = AudioSegment.from_mp3(includeDir + "scientist-3.mp3")
ST_4 = AudioSegment.from_mp3(includeDir + "scientist-4.mp3")
ST_5 = AudioSegment.from_mp3(includeDir + "scientist-5.mp3")
```

```python
filenames = glob.glob(includeDir+'*.mp3')
```

```python
filenameswithbeep = []
combined = AudioSegment.empty()


for filename in filenames:
  print("------",filename)
  audiofilename = AudioSegment.from_mp3(filename)
  filenameswithbeep.extend([audiofilename])
```

```python
for fname in filenameswithbeep:
    combined += fname
```

```python
combined.export(includeDir + "AUDIO_Scientist_Dutch_AUDIO_MAIN.mp3", format="mp3")
```
