```python
#X = beerCSV[["Temperature_Median","Temperature_Min","Temperature_Max","Rainfall","Weekend"]]
# ------------------------------ SAMPLE FOR SIMPLE LINEAR REGRESSION ------------------------------
X = beerCSV[["Temperature_Median"]].values
Y = beerCSV["Consumption_litres"].values
```

```python
print(X.shape)
print(Y.shape)
n = X.shape[0]
p = X.shape[1]
```

```python
# running the MOdel with MLFLOW

with mlflow.start_run():
  reg = LinearRegression()
  reg.fit(X,Y)
  R2 = reg.score(X,Y)
  Coefficients = reg.coef_
  Intercept = reg.intercept_
  adjusted_r2_score =  1 - (1-R2) * (n-1)/(n-p-1)

  mlflow.log_param("R2", R2)
  mlflow.log_param("Coefficients", Coefficients)
  mlflow.log_metric("Intercept", Intercept)
  mlflow.log_metric("Adjusted R2", adjusted_r2_score)
  #mlflow.log_metric("mae", mae)

  tracking_url_type_store = urlparse(mlflow.get_tracking_uri()).scheme

  # Model registry does not work with file store
  if tracking_url_type_store != "file":

      # Register the model
      # There are other ways to use the Model Registry, which depends on the use case,
      # please refer to the doc for more information:
      # https://mlflow.org/docs/latest/model-registry.html#api-workflow
      mlflow.sklearn.log_model(reg, "model", registered_model_name="Linear Regression")
  else:
      mlflow.sklearn.log_model(reg, "model")
```

```python
# FINDING THE ADJUSTED R2 FOR MULTIPLE LINEAR REGRESSION AS IT HELPS TO PENALISE VARIABLES AND CHOSE ONLY BEST VARIABLES TO GET BEST MODEL FIT

r2var = regMultiLR.score(X2,Y2)
n = X2.shape[0]
p = X2.shape[1]

adjusted_r2_score =  1 - (1-r2var) * (n-1)/(n-p-1)
print(adjusted_r2_score)
# we get 0.7187869447832966

# ADJUTED R2 SCORE = 0.7187869447832966
# R2 SCORE = 0.7226497614758338
# THUS HERE THE ADJUSTED R2 IS LOWER THEN THE R2, THUS ONE OR MORE OF THE INDEPENDENT VARIABLES HAVE LITTLE OF NOT USE FOR PREDICTING THE Y VARIABLE
```

```python
# FINDING WHICH  INDEPENDENT VARIABLE IS LESS SIGNIFICANT BASED ON ADJUSTED R2 AND HAS NO TO LITTLE USE FOR PREDITING Y
# LETS USE FEATURE SELECTION TO REMOVE VARIABLE
# FEATURE SELECTION SIMPLIFIES THE MODELS
# WE USE F_REGRESSION AVAIABLE IN FEATURE SELECTION CLASS TO FIND THE F STATISTICS and the P VALUES FOR EACH AND EVERY INDEPENDENT VARIABLE WITH THE DEPENDENT VARIABLE

# VARIABLES WHOSE P VALUE SIS LESS THEN 0.05 WILL BE DISREGARDED
# THUS IN OUR CASE IT WILL BE 
# F_REGRESSION FOR CONSUMPATION_LITRES < -  "Temperature_Median"
# F_REGRESSION FOR CONSUMPATION_LITRES < -  "Temperature_Min"
# F_REGRESSION FOR CONSUMPATION_LITRES < -  "Temperature_Max"
# F_REGRESSION FOR CONSUMPATION_LITRES < -  "Rainfall",
# F_REGRESSION FOR CONSUMPATION_LITRES < -  "Weekend"

from sklearn.feature_selection import f_regression

f_regression(X2,Y2)

# F STATISTICS = array([178.93829777,  66.11002412, 255.42804139,  14.16337347,124.9139062 ])
# P VALUES = array([1.86674745e-33, 6.81355233e-15, 6.54772593e-44, 1.95381942e-04,4.00796880e-25])

pValVar = f_regression(X2,Y2)[1]
print(pValVar)
print(pValVar.round(3))
#          ["Temperature_Median","Temperature_Min","Temperature_Max","Rainfall",      "Weekend"]
# P VAL is [ 1.86674745e-33       6.81355233e-15    6.54772593e-44    1.95381942e-04     4.00796880e-25]
# AFTER ROUND OF  [0. 0. 0. 0. 0.]

# THUS WITH ALL THE P VALUES CALCULATED WE CAN SEE THAT ALL THE VARIABLES ARE USE FUL  IN PREDICTING THE Y AS ALL VALUES ARE LESS THEN 0.05 WHICH ARE 
# SIGNIFICANT TO PREDICTING THE Y VARIABLE
```

```python
#ABOVE FINDING COLLINERITY USING CORRELATION MATRIX FOUND THAT
#1.  TEMP_MIN, TEMP_MAX, TEMP_MEDIAN ARE CORRELATED
#2. ETS VERIFY THE SAME USING VIF
```

```python
VIFdataFrame =  pd.DataFrame()
VIFdataFrame["Columns/Features"] = X.columns
VIFdataFrame["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
#(X.shape[1]) = considering all columns available in the X Array
print(VIFdataFrame)
# WE SEE THAT VIF VALUES FOR TEMP_MED, TEMP_MIN, TEMP_MAX ARE VERY LARGE THUS THERE EXISTS COLLINEARITY BETWEEN THE PREDICTORS, 
# WHICH IN TURN WILL INCREASE THE ERROR DURING PREDICTION, AND WE WILL NOT KNOW WHICH VARIABLE GETS THE BEST FIT LINE FOR THE RESPONSE.
```

```python
# Remove variables and KEEP GENERATING THE VIF SCORE UNTIL THE VALUES FALL BELOW 1 IN VIF FACTOR SCORE
# HERE WE EXPERIMENETED WITH VARIABLE  -------------  Temperature_Median
def check_vif_drop_column(X,column_name): 
 X = X.drop(columns=column_name) 
 vif = pd.DataFrame() 
 vif["Columns/Features"] = X.columns
 vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
 return vif,X

vifUpdated,X = check_vif_drop_column(X,"Temperature_Median")
print(vifUpdated)
```

```python
# Remove variables and KEEP GENERATING THE VIF SCORE UNTIL THE VALUES FALL BELOW 1 IN VIF FACTOR SCORE
# HERE WE EXPERIMENETED WITH VARIABLE  -------------  Temperature_Min
def check_vif_drop_column(X,column_name): 
 X = X.drop(columns=column_name) 
 vif = pd.DataFrame() 
 vif["Columns/Features"] = X.columns
 vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
 return vif,X

vifUpdated2,X = check_vif_drop_column(X,"Temperature_Min")
print(vifUpdated2)
```

```python
def split_train_data(X,Y):
 X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30)
 return(X_train, X_test, Y_train, Y_test)
X_train, X_test, Y_train, Y_test = split_train_data(X,Y)
```

```python
# lets build linear reg model
def model_fit(model,X_train, Y_train):
 model = LinearRegression() 
 model.fit(X_train, Y_train)
 return model
lin_model = model_fit(LinearRegression,X_train, Y_train)
```

```python
#PREDICTING WITH TEST DATA


def scores_(model,X,Y):
 y_predict = model.predict(X)
 rmse = (np.sqrt(mean_squared_error(Y, y_predict)))
 r2 = r2_score(Y, y_predict)
 print("RMSE is {}".format(rmse))
 print("R2 score is {}".format(r2))
print("The model performance of training set")
scores_(lin_model,X_train,Y_train)
print("--------------------------------------")
print("The model performance of testing set")
scores_(lin_model,X_test,Y_test)
```
